```python
gl.glClear(gl.GL_COLOR_BUFFER_BIT)
#Starts From Here
def EightWay_midpoint(f_cordinate,s_cordinate):
  x_1,y_1 = f_cordinate
  x_2,y_2 = s_cordinate
  dx = x_2 - x_1
  dy = y_2 - y_1
  #ZOne finding and converting to Zone 0
  cordinate_1,cordinate_2,zone= f_cordinate,s_cordinate,0
  if abs(dx)>= abs(dy):
    if dx>=0 and dy>=0:
      #zone 0
      cordinate_1,cordinate_2,zone= f_cordinate,s_cordinate,0
    elif dx<=0 and dy>=0:
      #zone 3
      cordinate_1,cordinate_2,zone =(-x_1,y_1),(-x_2 ,y_2),3
    elif dx<=0 and dy<=0:
      #zone 4
      cordinate_1,cordinate_2,zone= (-x_1,-y_1),(-x_2 ,-y_2),4
    elif dx>=0 and dy<=0:
      #zone 7
      cordinate_1,cordinate_2,zone =(x_1,-y_1),(x_2 ,-y_2),7
  elif abs(dx)<= abs(dy):
    if dx>=0 and dy>=0:#zone1
      cordinate_1,cordinate_2,zone= (y_1,x_1),(y_2,x_2),1
    elif dx<=0 and dy>=0:#zone 2
      cordinate_1,cordinate_2,zone =(y_1,-x_1),(y_2,-x_2),2
    elif dx<=0 and dy<=0:
      #zone 5
      cordinate_1,cordinate_2,zone =(-y_1,-x_1),(-y_2,-x_2),5
    elif dx>=0 and dy<=0:
      #zone 6
      cordinate_1,cordinate_2,zone= (-y_1,x_1),(-y_2,x_2),6

  cordinates= []
  #Applying Mid point
  x_1,y_1 = cordinate_1
  x_2,y_2 = cordinate_2
  dx = x_2 - x_1
  dy = y_2 - y_1
  d_init = 2*dy - dx
  d = d_init
  incrE, incrNE = 2 * dy,2 * (dy - dx)
  temp_x,temp_y = x_1,y_1
  while temp_x<=x_2:
    cordinates.append((temp_x,temp_y))
    # print(cordinates)
    if d>=0: #NE
      d += incrNE
      temp_x += 1
      temp_y += 1
    elif d<0:
      d += incrE
      temp_x += 1
  #Convert to original zone
  new_cordinates = []
  if zone == 0:
    new_cordinates=cordinates
  elif zone == 1:
    for item in cordinates:
      #item = (3,5)
      x,y = item
      new_cordinates.append((y,x))
  elif zone == 2:
    for item in cordinates:
      x,y = item
      new_cordinates.append((-y,x))
  elif zone == 3:
    for item in cordinates:
      x,y = item
      new_cordinates.append((-x,y))
  elif zone == 4:
    for item in cordinates:
      x,y = item
      new_cordinates.append((-x,-y))
  elif zone == 5:
    for item in cordinates:
      x,y = item
      new_cordinates.append((-y,-x))
  elif zone == 6:
    for item in cordinates:
      x,y = item
      new_cordinates.append((y,-x))
  elif zone == 7:
    for item in cordinates:
      x,y = item
      new_cordinates.append((x,-y))

  gl.glColor3f(255, 255, 0)
  gl.glPointSize(5)
  gl.glBegin(gl.GL_LINES)
  for item in new_cordinates:
    #item = (3,5)
    ordinate,abscissa = item
    gl.glVertex2f(ordinate/(WIDTH/2), abscissa/(HEIGHT/2))
  gl.glEnd()
    
user_num=input("Enter Last 2 digits:")
f = 0
for i in range(len(user_num)-2,len(user_num)):
  #20201234
  digit = int(user_num[i])
  offset = f*30
  if digit==1:
    a = -10 + offset
    b_y= 70
    e_y = 0
    EightWay_midpoint((a,b_y),(a,e_y)) #(3,4),(10,20)
    EightWay_midpoint((a,b_y),(a-10,b_y-10))
    EightWay_midpoint((a-10,e_y),(a+10,e_y))
  elif digit==2:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b))
    EightWay_midpoint((c,b),(c,d_y))
    EightWay_midpoint((c,d_y),(a,d_y))
    EightWay_midpoint((a,d_y),(a,c_y))
    EightWay_midpoint((a,c_y),(c,c_y))
  elif digit== 0:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b))
    EightWay_midpoint((a,b),(a,c_y))
    EightWay_midpoint((a,c_y),(c,c_y))

    EightWay_midpoint((c,c_y),(c,b))
  elif digit== 8:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b))
    EightWay_midpoint((a,b),(a,c_y))
    EightWay_midpoint((a,c_y),(c,c_y))
    EightWay_midpoint((a,d_y),(c,d_y))
    EightWay_midpoint((c,c_y),(c,b))
  elif digit== 9:
    a = -10 + offset
    b= 70
    b_x= 70 + offset 
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35 
    EightWay_midpoint((a,b),(c,b)) # 0
    EightWay_midpoint((c,b),(c,d_y)) # 1
    EightWay_midpoint((a,d_y),(c,d_y)) # 2
    EightWay_midpoint((a,d_y),(a,b)) #3
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
    EightWay_midpoint((c,c_y),(a,c_y)) #5
  elif digit== 4:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((c,b),(c,d_y)) # 1
    EightWay_midpoint((a,d_y),(c,d_y)) # 2
    EightWay_midpoint((a,d_y),(a,b)) #3
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
  elif digit== 5:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b)) # 0
    EightWay_midpoint((a,d_y),(c,d_y)) # 2
    EightWay_midpoint((a,d_y),(a,b)) #3
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
    EightWay_midpoint((c,c_y),(a,c_y)) #5
  elif digit== 6:
    a = -10 + offset
    b= 70
    b_x= 70 + offset 
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b)) # 0
    EightWay_midpoint((a,d_y),(c,d_y)) # 2
    EightWay_midpoint((a,d_y),(a,b)) #3
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
    EightWay_midpoint((c,c_y),(a,c_y)) #5
    EightWay_midpoint((a,d_y),(a,c_y)) #6
  elif digit== 7:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b)) # 0
    EightWay_midpoint((c,b),(c,d_y)) # 1
    EightWay_midpoint((a,d_y),(a,b)) #3
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
  elif digit== 3:
    a = -10 + offset
    b= 70
    b_x= 70 + offset
    c = 0 + offset
    c_y = 0
    d = 35 + offset
    d_y = 35
    EightWay_midpoint((a,b),(c,b)) # 0
    EightWay_midpoint((c,b),(c,d_y)) # 1
    EightWay_midpoint((a,d_y),(c,d_y)) # 2
    EightWay_midpoint((c,d_y),(c,c_y)) # 4
    EightWay_midpoint((c,c_y),(a,c_y)) #5
  f+=1


img_buf = gl.glReadPixelsub(0, 0, WIDTH, HEIGHT, gl.GL_RGB, gl.GL_UNSIGNED_BYTE)
img = np.frombuffer(img_buf, np.uint8).reshape(HEIGHT, WIDTH, 3)[::-1]
show.image(img/255.0)
```
